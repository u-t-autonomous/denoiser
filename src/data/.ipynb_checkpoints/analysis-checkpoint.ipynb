```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from math import sqrt

def t_test(mean_1, std_1, size_1, mean_2, std_2, size_2):
    return abs((mean_1 - mean_2) / sqrt((std_1**2/size_1) + (std_2**2/size_2)))

wo_correction = pd.read_csv("no_correction_err_3_bp_5_2D.csv")
w_correction_5 = pd.read_csv("swap_correction_err_3_bp_5_2D.csv")
# w_correction_4 = pd.read_csv("with_correction_beta_4_err_3.csv")
# w_correction_6 = pd.read_csv("with_correction_beta_6_err_3.csv")
# w_correction_8 = pd.read_csv("with_correction_beta_8_err_3.csv")
wo_correction = wo_correction[pd.notnull(wo_correction['ts_95'])]
w_correction_5 = w_correction_5[pd.notnull(w_correction_5['ts_95'])]
# w_correction_8 = w_correction_8[pd.notnull(w_correction_8['ts_95'])]
colors = ['b', 'c', 'y', 'm', 'r']
#w2=plt.scatter(w_correction_2['er_95'],w_correction_2['ts_95'],color=colors[1], marker='v')
#w4=plt.scatter(w_correction_4['er_95'],w_correction_4['ts_95'],color=colors[3], marker='v')
#w6=plt.scatter(w_correction_6['er_95'],w_correction_6['ts_95'],color=colors[2], marker='v')
#w8=plt.scatter(w_correction_8['er_95'],w_correction_8['ts_95'],color=colors[4], marker='v')
#wo=plt.scatter(wo_correction['er_95'],wo_correction['ts_95'],color=colors[0], marker='o', alpha=.3)

x = w_correction_5['er_95']
y = w_correction_5['ts_95']

z = np.polyfit(x, y, 2)
f = np.poly1d(z)

# calculate new x's and y's
x_new = np.linspace(min(x.values), max(x.values), 50)
y_new = f(x_new)

#t, m, b = np.polyfit(x, y,2)
w5=plt.scatter(w_correction_5['er_95'],w_correction_5['ts_95'],color=colors[0], marker='v')
plt.plot(x_new, y_new, '-',color=colors[0])

# x = w_correction_4['er_95']
# y = w_correction_4['ts_95']
# m, b = np.polyfit(x, y, 1)
# w4=plt.scatter(w_correction_4['er_95'],w_correction_4['ts_95'],color=colors[1], marker='v')
# plt.plot(x, m*x + b, '-',color=colors[1])

# x = w_correction_6['er_95']
# y = w_correction_6['ts_95']
# m, b = np.polyfit(x, y, 1)
# w6=plt.scatter(w_correction_6['er_95'],w_correction_6['ts_95'],color=colors[2], marker='v')
# plt.plot(x, m*x + b, '-',color=colors[2])

# x = w_correction_8['er_95']
# y = w_correction_8['ts_95']
# m, b = np.polyfit(x, y, 1)
# w8=plt.scatter(w_correction_8['er_95'],w_correction_8['ts_95'],color=colors[3], marker='v')
# plt.plot(x, m*x + b, '-',color=colors[3])

x = wo_correction['er_95']
y = wo_correction['ts_95']
#m, b = np.polyfit(x, y, 1)

z = np.polyfit(x, y, 2)
f = np.poly1d(z)

# calculate new x's and y's
x_new = np.linspace(min(x.values), max(x.values), 50)
y_new = f(x_new)

wo=plt.scatter(wo_correction['er_95'],wo_correction['ts_95'],color=colors[4], marker='v')
plt.plot(x_new, y_new, '-',color=colors[4])



plt.legend((w5,wo),
           ('beta = .5','No correction'),
           scatterpoints=1,
           loc='top right',
           ncol=3,
           fontsize=12)
plt.ylabel('Number of steps')
plt.xlabel('Error rate (%)')
plt.show()
```

```python
df = pd.DataFrame()
df['with beta .5'] = w_correction_5['ts_95']
# df['with beta .6'] = w_correction_6['ts_95']
# df['with beta .4'] = w_correction_4['ts_95']
# df['with beta .2'] = w_correction_2['ts_95']
df['without'] = wo_correction['ts_95']
print df.describe()
df.hist()
plt.show()
print 'T value for beta .5 = ', t_test(df['with beta .5'].mean(), df['with beta .5'].std(), df['with beta .5'].size, df['without'].mean(), df['without'].std(), df['without'].size)
# print 'T value for beta .4 = ', t_test(df['with beta .4'].mean(), df['with beta .4'].std(), 1000, df['without'].mean(), df['without'].std(), 1000)
# print 'T value for beta .6 = ', t_test(df['with beta .6'].mean(), df['with beta .6'].std(), 1000, df['without'].mean(), df['without'].std(), 1000)
# print 'T value for beta .8 = ', t_test(df['with beta .8'].mean(), df['with beta .8'].std(), 1000, df['without'].mean(), df['without'].std(), 1000)
```
